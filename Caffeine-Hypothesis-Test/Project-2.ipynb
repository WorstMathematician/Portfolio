# Coffee Caffeine Content Hypothesis Test

This notebook performs a one-sample Z-test to evaluate whether the average caffeine content differs from the advertised value of 120 mg.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Sample caffeine contents in milligrams
data = np.array([
    118, 122, 125, 117, 119, 121, 123, 116, 124, 120,
    118, 121, 126, 115, 117, 123, 122, 119, 124, 118,
    116, 120, 123, 127, 121, 122, 119, 124, 125, 118,
    117, 123, 121, 120, 116, 122, 123, 125, 118, 121,
    119, 120, 117, 124, 122, 118, 121, 123, 120, 116
])

mu0 = 120      # claimed population mean
sigma = 15    # known population standard deviation
alpha = 0.05  # significance level
```

```python
n = len(data)
sample_mean = data.mean()
z_stat = (sample_mean - mu0) / (sigma / np.sqrt(n))
critical_value = norm.ppf(1 - alpha / 2)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
decision = 'Reject H0' if abs(z_stat) > critical_value else 'Fail to Reject H0'

print(f'Sample size: {n}')
print(f'Sample mean: {sample_mean:.4f} mg')
print(f'Z-statistic: {z_stat:.4f}')
print(f'Critical z-value: ±{critical_value:.4f}')
print(f'P-value: {p_value:.4f}')
print(f'Decision: {decision}')
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data, bins=10, edgecolor='black')
plt.axvline(sample_mean, linestyle='--', linewidth=2, label=f'Sample mean = {sample_mean:.2f}')
plt.axvline(mu0, linestyle=':', linewidth=2, label=f'Claimed mean = {mu0}')
plt.title('Histogram of Coffee Caffeine Content')
plt.xlabel('Caffeine content (mg)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## Conclusion

Because the p-value is greater than 0.05 and the Z-statistic is within the critical bounds, we fail to reject the null hypothesis. The sample does not provide strong evidence that the true mean caffeine content differs from 120 mg.
