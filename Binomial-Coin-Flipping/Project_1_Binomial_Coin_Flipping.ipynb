# Bayesian Binomial Coin Flipping

This notebook estimates the probability of heads using a Beta-Binomial Bayesian update.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
# Prior assumptions
alpha_prior = 2
beta_prior = 2

# Simulated coin flip experiment
np.random.seed(42)
true_theta = 0.671
n_flips = 100
flips = np.random.binomial(1, true_theta, size=n_flips)
heads_count = int(flips.sum())
tails_count = int(n_flips - heads_count)
observed_proportion = heads_count / n_flips

print(f'Heads: {heads_count}')
print(f'Tails: {tails_count}')
print(f'Observed proportion of heads: {observed_proportion:.3f}')
```

```python
# Posterior update: theta | X ~ Beta(alpha + heads, beta + tails)
alpha_post = alpha_prior + heads_count
beta_post = beta_prior + tails_count
posterior_mean = alpha_post / (alpha_post + beta_post)

print(f'Posterior Distribution: Beta({alpha_post}, {beta_post})')
print(f'Posterior Mean: {posterior_mean:.3f}')
```

```python
theta = np.linspace(0, 1, 1000)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(theta, beta.pdf(theta, alpha_prior, beta_prior), linewidth=2)
plt.title(f'Prior Density: Beta({alpha_prior}, {beta_prior})')
plt.xlabel('theta: probability of heads')
plt.ylabel('Density')

plt.subplot(1, 2, 2)
plt.plot(theta, beta.pdf(theta, alpha_post, beta_post), linewidth=2, label='Posterior Density')
plt.axvline(posterior_mean, linestyle='--', label=f'Posterior Mean ({posterior_mean:.3f})')
plt.title(f'Posterior Density: Beta({alpha_post}, {beta_post})')
plt.xlabel('theta: probability of heads')
plt.ylabel('Density')
plt.legend()

plt.tight_layout()
plt.show()
```

## Interpretation

The Beta(2, 2) prior is centered near 0.5 and represents a mild belief that the coin may be fair. After observing 69 heads in 100 simulated flips, the posterior updates to Beta(71, 33), with a posterior mean near 0.683. This suggests evidence of a bias toward heads in the simulated experiment.
